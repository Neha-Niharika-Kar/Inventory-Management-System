```python
record = json.loads(text)
record
```

```python
while True:
    print('\nEnter new product details in order: ')
    new = input().split(' ')
    
    if new[0] in record.keys():
        record[new[0]]['quantity'] += int(new[3])
    else:
        update = {}
        update['name'] = new[1] 
        update['price'] = float(new[2])
        update['quantity'] = int(new[3])
        update['category'] = new[4]
        update['discount'] = int(new[5])
        record[new[0]] = update

    ch = input('\nDo you want to add another product? ')
    if ch == 'Y' or ch == 'y':
        continue
    else:
        break
```

```python
add = json.dumps(record)
fl = open("records.json", 'w')
fl.write(add)
fl.close()
```
