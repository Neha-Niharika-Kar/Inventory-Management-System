# **Purchasing Products**

```python
import json
```

```python
f = open("records.json", 'r')
text = f.read()
f.close()
```

```python
record = json.loads(text)
```

```python
print("Products Available: \n")
record
```

```python
s = 1
sales = {}
while True:
    choice = int(input('\nEnter your choice: '))

    if choice == 1:
        user = input('What is your name? ')
        n = input('Enter the item key and quantity you want to buy: ')
        prod = {}
        m = n.split(' ')
        prod[m[0]] = int(m[1])
    
        i = m[0]
        if i in record.keys():
            if record[i]['quantity'] >= prod[i]:
                print('Cost of {} {} = {}'.format(prod[i], record[i]['name'], prod[i]*record[i]['price']))
                print('Discount = {} %'.format(record[i]['discount']))
                amt = prod[i]*record[i]['price']*(1.0 - record[i]['discount']/100.0)
                print('Total amount =', amt)
                record[i]['quantity'] -= prod[i]

                item = {}
                item['customer'] = user
                item['name'] = record[i]['name']
                item['quantity'] = prod[i]
                item['price'] = record[i]['price']
                item['SP/item'] = record[i]['price']*(1.0 - record[i]['discount']/100.0)
                item['total'] = item['SP/item']*prod[i]
                sales[s] = item
                s += 1

            else: 
                print("Required quantity not available!")
                continue
  
    if choice == 2:
        print("Thank you for shopping!")
        break
```

```python
fd = open("sales.json", 'w')
fd.write(json.dumps(sales))
fd.close()
```
